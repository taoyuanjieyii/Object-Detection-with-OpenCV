#### In this Project you will learn Object Detection with OpenCV and Python. You will detect faces, eyes and both of them from Images, people walking and cars moving from Videos and car's plate from Images.

```python
import os
from os import getcwd
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Create Classifier

```python
path_eye = f"{getcwd()}/Haarcascade/haarcascade_eye.xml"    #path to dataset

eye_classifier = cv2.CascadeClassifier(path_eye)                     
```

#### Open Image

```python
image = cv2.imread('dataset/Images/IMG_201.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

#### Detect Faces

```python
eye = eye_classifier.detectMultiScale(fix_img, 1.3, 5)

if eye is ():
    print("No eyes found")
```

```python
def detect_eyes(fix_img):
    eye_rects = eye_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(fix_img,
                     (x,y),          # upleft corner
                     (x+w, y+h),     # downright corner
                     (255,255,255),  # white color for rectangles
                     10)             # 
    return fix_img


result = detect_eyes(fix_img)
plt.imshow(result)
    
```
