#### In this Project you will learn Object Detection with OpenCV and Python. You will detect faces, eyes and both of them from Images, people walking and cars moving from Videos and car's plate from Images.

```python
import os
from os import getcwd
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Create Classifier

```python
path_frontalface = f"{getcwd()}/Haarcascade/haarcascade_frontalface_default.xml"    #path to dataset

face_classifier = cv2.CascadeClassifier(path_frontalface)                     
```

#### Open Image

```python
image = cv2.imread('dataset/Images/IMG_201.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

#### Detect Faces

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)

if faces is ():
    print("No faces found")
```

```python
def detect_face(fix_img):
    
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h)  in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w, y+h),
                     (255,0,0),
                     10)
    return fix_img


result = detect_face(fix_img)
plt.imshow(result)
    
```
