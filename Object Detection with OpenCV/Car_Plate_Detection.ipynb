#### In this Project you will learn Object Detection with OpenCV and Python. You will detect car's plate from Images.

```python
import os
from os import getcwd
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('dataset/Images/car6.JPG')
```

```python
def display(img):
    fig = plt.figure(figsize = (10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

#### Create Classifier

```python
path_plate = f"{getcwd()}/Haarcascade/haarcascade_russian_plate_number.xml"   

plate_classifier = cv2.CascadeClassifier(path_plate)                     
```

#### Capture Video

```python
def detect_plate(img):
    
    plate_img = img.copy()
    
    plate_rects = plate_classifier.detectMultiScale(plate_img, 1.1, 1)
    
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(plate_img,
                     (x,y),
                     (x+w, y+h),
                     (255, 225, 225),
                     5)
            
    return plate_img
```

```python
result = detect_plate(img)
display(result)
```
